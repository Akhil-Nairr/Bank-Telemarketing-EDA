```python

# first converting all time durations to seconds
def convert_to_seconds(time_str):
    if 'sec' in time_str:
        return int(time_str.replace(' sec', ''))
    elif 'min' in time_str:
        # Convert minutes to seconds
        return int(float(time_str.replace(' min', '')) * 60)
    else:
        return np.nan 

df['duration'] = df['duration'].apply(convert_to_seconds)
```

```python
# Calculate the mean of the 'age' column and fill NaN values with this mean
mean_age = df['age'].mean()
df['age'] = df['age'].fillna(mean_age)

# Convert 'age' from float to integer
df['age'] = df['age'].astype(int)
```

```python
df.isnull().sum()
```

```python
mode_month = df['month'].mode()[0]
df['month'] = df['month'].fillna(mode_month)

df['day'] = df['day'].astype(str)
df['month'] = df['month'].str.extract('([a-zA-Z]+)', expand=False)

#  convert to datetime
fixed_year = "2017"
df['date'] = pd.to_datetime(df['day'] + '-' + df['month'] + '-' + fixed_year, format='%d-%b-%Y')
```

```python
df[['job', 'education']] = df['jobedu'].str.split(',', expand=True)
```

```python


# lets caluculate the mean of the 'age' column and fill NaN values with this mean
mean_age = df['age'].mean()
df['age'] = df['age'].fillna(mean_age).astype(int)  # Fill missing and convert to integer

# lets calculate the mode for the 'response' column and fill NaN values
mode_response = df['response'].mode()[0]
df['response'] = df['response'].fillna(mode_response)
df.isnull().sum()
```

```python
# Function to detect outliers using the Interquartile Range (IQR)
def detect_outliers(data, feature):
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(data[feature] < lower_bound) | (df[feature] > upper_bound)]
    return outliers

# Apply the outlier detection function to numerical columns
outliers_age = detect_outliers(df, 'age')
outliers_salary = detect_outliers(df, 'salary')
outliers_balance = detect_outliers(df, 'balance')
outliers_call_duration = detect_outliers(df, 'duration')

# Summarize the number of outliers for each feature
outlier_summary = {
    'age': outliers_age.shape[0],
    'salary': outliers_salary.shape[0],
    'balance': outliers_balance.shape[0],
    'duration': outliers_call_duration.shape[0]
}

# Print the summary of outliers
print("Outlier Summary:", outlier_summary)
```

```python
numerical_independent_variables = df.select_dtypes(include=['int64', 'float64'])

correlation_matrix = numerical_independent_variables.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Independent Variables')
plt.show()
```

```python
# Set a threshold for high correlation
high_correlation_threshold = 0.7

# Find highly correlated features
high_corr_features = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > high_correlation_threshold:
            high_corr_features.append((correlation_matrix.columns[i], correlation_matrix.columns[j]))

# Print highly correlated features
print("Highly Correlated Features:")
for feature_pair in high_corr_features:
    print(feature_pair)
```

```python

def identify_outliers(column):
    Q1 = np.percentile(column, 25)
    Q3 = np.percentile(column, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = column[(column < lower_bound) | (column > upper_bound)]
    return outliers

# Identifying outliers for each numerical column
for column in numerical_columns:
    outliers = identify_outliers(df[column])
    print(f"Outliers in {column}:")
    print(outliers)
    print()
```

```python
def winsorize(column, lower_percentile, upper_percentile):
    lower_bound = np.percentile(column, lower_percentile)
    upper_bound = np.percentile(column, upper_percentile)
    winsorized_column = np.clip(column, lower_bound, upper_bound)
    return winsorized_column

for column in numerical_columns:
    df[column] = winsorize(df[column], 5, 95)
```
